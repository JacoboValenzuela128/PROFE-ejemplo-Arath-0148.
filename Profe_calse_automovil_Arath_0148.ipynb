#### Clase automovil con 3 atributos funciones: leerdatos(), Mostrardatos()

####  Que sea interactivo --> los input
#### Crear 3 objetos: carro1, carro2 y carro3.
#### Nos usar while true, if, for, listas ni diccionarios
![texto del vínculo](https://raw.githubusercontent.com/JacoboValenzuela128/PROFE-ejemplo-Arath-0148./refs/heads/main/infoclaseauto.png)

```python
class Automovil:
  def __init__(self, marca, placa, precio):
      self.marca = marca
      self.placa = placa
      self.precio = precio

  def leerdatos(self):
      self.marca = input("Ingrese la marca: ")
      self.placa = input("Ingrese la placa: ")
      self.precio = input("Ingrese el precio: ")
      return self.marca, self.placa, self.precio

  def Mostrardatos(self):
      print("Marca: ", self.marca)
      print("Placa: ", self.placa)
      print("Precio: ", self.precio)
      return self.marca, self.placa, self.precio

print("Carro N1")
carro1 = Automovil("", "", "")
carro1.leerdatos()
carro1.Mostrardatos()
print("-------------------------------------------------------------------")
print("Carro N2")
carro2 = Automovil("", "", "")
carro2.leerdatos()
carro2.Mostrardatos()
print("-------------------------------------------------------------------")
print("Carro N3")
carro3 = Automovil("", "", "")
carro3.leerdatos()
carro3.Mostrardatos()
print("-------------------------------------------------------------------")
print("Fin del programa. Software desarrollado por Arath VAlenzuela NL52.")
```
